```python
df.loc[:, 'date'] = df.loc[:, '日付'].apply(pd.to_datetime)
```

```python
df.loc[:, "date"]
```

```python
df.loc[:, "摂取カロリー"] = df.loc[:, "摂取カロリー"].astype(np.float32)
```

```python
df = df.set_index("date")
```

```python
df.head()
```

```python
df.sort_values(by="歩数")
```

```python
df.sort_values(by="歩数", ascending=False).head()
```

```python
df = df.drop("日付", axis=1)
```

```python
df.tail()
```

```python
df.loc[:, "歩数/カロリー"] = df.loc[:, "歩数"] / df.loc[:, "摂取カロリー"]
df
```

```python
def exercise_judge(ex):
    if ex <= 3.0:
        return "Low"
    elif 3.0 < ex <= 6.0:
        return "Middle"
    else:
        return "High"
```

```python
df.loc[:, "運動指数"] = df.loc[:, "歩数/カロリー"].apply(exercise_judge)
df
```

```python
df.to_pickle("notebooks/data/df_201704health.pickle")
```

```python
df_moved = pd.get_dummies(df.loc[:, "運動指数"], prefix="運動")
df_moved
```

```python
df_moved.to_pickle("notebooks/data/df_201704moved.pickle")
```

```python
# 22
import pandas as pd 
df = pd.DataFrame([[40, "a", True],[20, "b", False],[30, "c", False]]) 
df.index = ["01", "02", "03"] 
df.columns = ["A", "B", "C"] 

print(df)

def judge(arg): 
    if arg < 50: 
        return "low" 
    elif arg < 70: 
        return "middle" 
    else: 
        return "high" 
  
df.loc[:, "C"] = df.iloc[:, 0] * 2 

print(df)

df.loc[:, "B"] = df.iloc[:, 2].apply(judge) 

print(df)

_ = df["C"] > 50 
df = df[_] 
 
print(df)
    
print(df.iloc[0 , 2], df.loc["03" ,"B"])
```

```python
# 24
#DataFrameの各要素が欠損値かどうかを確かめるメソッドは、isnullである。
#欠損値が存在する行や列を削除するメソッドは、dropnaである。
"fillnaメソッドの引数にmethod='ffill'を与えると、欠損値を1つ前方の値で補完する。
"mean()メソッドの結果をfillnaメソッドに引数として渡すと、欠損値を他の値の平均値で補完できる。
"mode().iloc[1.:]の結果をfillnaメソッドに引数として渡すと、欠損値を最頻値で補完できる。
```
